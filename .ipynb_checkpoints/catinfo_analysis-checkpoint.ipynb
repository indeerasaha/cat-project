```python
import pandas as pd 
import numpy as np 
```

```python
import csv

# Open the CSV file
with open('cat_information.csv', mode='r') as file:
    csv_reader = csv.reader(file)
    
    full_list = list(csv_reader)
    
categories = {}

values = full_list[0][5:19]
for category in values:
    categories.update({category: len(category)})
    
full_list.pop(0)

dict_list = []
age_class = ['Baby', 'Adult', 'Young', 'Senior']
sizes = ['Medium', 'Large', 'X-Large', 'Small']
for cat in full_list:
    #cat is an indexable list 
    entry = {}
    entry['Name'] = cat[0]
    entry['Breed'] = cat[1]
    entry['Gender'] = cat[2]
    if cat[3] in age_class: 
        entry['Age_class'] = cat[3]
    elif cat[3] in sizes:
        entry['Size'] = cat[3]
    if (cat[3] == 'Status:Adopted!' or cat[3] == 'Status: Available for Adoption (adoption info)'):
        entry['Size'] = np.nan 
        entry['Age_class'] = np.nan
        
    i = 3
    while(i < len(cat)):
        
        lst = cat[i].split(':')
        if len(lst) == 2: 
            entry[lst[0]] = lst[1] 
     
        i+=1 
    dict_list.append(entry)
            
print(dict_list)
```

```python
cat_df = pd.DataFrame(dict_list)

cat_df
```

```python
#basic info of the dataset 

print(f"cat_df has {cat_df.shape[0]} rows and {cat_df.shape[1]} columns")
print(cat_df.columns)
```

```python
#converting the breed column into Breed and Coat type 

cat_df['Coat_Qualities'] = cat_df['Breed'].str.split('/').str[-1]
```

```python
cat_df['Breed_Type'] = cat_df['Breed'].str.split('/').str[0].str.strip()
cat_df['Coat_Spread'] = cat_df['Breed'].str.split('/').str[-1].str.split('(').str[0].str.strip()
#coat spread takes the last value separated by the '/'
```

```python
pd.set_option('display.max_rows', 100)
#cat_df['Breed_Type'].value_counts() 
```

```python
#extracting coat length from coat qualities 
cat_df['Coat_Length'] = cat_df['Coat_Qualities'].str.extract(r'\((.*?)\)')
```

```python
#cat_df['Coat_Spread'].value_counts()
```

```python
import re 
cat_df['Breed_Type'] = cat_df['Breed_Type'].str.replace(r'\(.*?\)', '', regex=True).str.strip()
#cleansing any breed types that still include coat length 
```

```python
cat_df['Coat_Length'].value_counts()
```

```python
cat_df['Slash_Count'] = cat_df['Breed'].str.count('/')
#calculating sub parts for each cat's breed + coat color description 
```

```python
mask = cat_df[cat_df['Slash_Count'] >= 2].index

# function extracts mixed breed types 
def extract_mixed_breeds(breed_type):
    parts = breed_type.split(' / ')
    return ' / '.join(parts[:2])

#applies to all cats that have more than one breed type listed 
cat_df.loc[mask, 'Breed_Type'] = cat_df.loc[mask, 'Breed'].apply(extract_mixed_breeds)
```

```python
#checking the values in every column 

for col in cat_df.columns:
    print(cat_df[col].value_counts())
    
```
